Import library (collections of pre-written functions that extend the capabilities of the Python programming language)

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

Create the data set using the function DataFrame

```python
stocks=['AAPL','ABT','ADBE','AMZN','AVGO','BAC','BRK-B','CMCSA','CRM','CSCO']
data=pd.DataFrame()
```

Plotting the historical data

```python
for i in stocks:
  data[i]=yf.download(i, start='2020-12-01', end='2024-05-31')['Adj Close']

display(data)
```

We graph stock price trends over time

Then calculate the log-returns of the stocks, and delate the NaN observations

```python
r=np.log(data/data.shift()).dropna()
mu=r.mean()*252 # We create the vector of annualised average stocks returns

display(r)
display(mu)
```

Now can calculate the variance, standard deviation and covariance of the returns using the functions var, sdt and cov.

```python
variance=r.var()
standard_deviation=r.std()
cov=r.cov()
sigma=r.cov()*252 #annualised variance-covariance matrix


display(variance)
display(standard_deviation)
display(cov)
display(sigma)
```

```python
correlation=r.corr()

display(correlation)
```
