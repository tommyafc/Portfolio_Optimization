```python
import pandas as pd #data analysis
import numpy as np #working with arrays
import yfinance as yf #download stocks data from yahoo finance
import matplotlib.pyplot as plt #plot
```

```python
#create dataset

stocks=['AAPL','BTC-USD','ERG.MI','CL=F','^DJI','CPR.MI','^GSPC']

#A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

data=pd.DataFrame()

for i in stocks:
  data[i]=yf.download(i, start = '2020-12-01', end='2024-01-01')['Adj Close']

print(data)
display(data)
```

```python
for i in stocks:
  data[i].plot(color='k')
  plt.title("historical data "+ i)
  plt.xlabel("Data")
  plt.ylabel("Prezzo")
  plt.show()
```

```python
#return of the assets

r=np.log(data/data.shift()).dropna()
mu=r.mean()*256

display(r)
display(mu)
```

```python
display(r.mean()) #mean
display(r.var()) #variance
display(r.std()) #standard deviation
```

```python
sigma=r.cov()*256 #covariance -> var/cov matrix
display(sigma)
```

```python
#correlation matrix
display(r.corr())
```

```python
#optimization portfolio library

!pip install PyPortfolioOpt
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
from pypfopt import plotting
```

```python
#bounds (0,1) or (-1,1)

ef = EfficientFrontier(mu,sigma, weight_bounds=(0 , 1))
fig, ax = plt.subplots()
plotting.plot_efficient_frontier(ef, ax=ax, show_assets=True)

ef.efficient_return(-1.0)
ret_min, std_min, _ = ef.portfolio_performance()
ax.scatter(std_min, ret_min, marker=".", s=100, c="r", label='min var')

plt.show()
```

```python
weights = ef.efficient_return(-1.0)  # find the portfolio min variance

cleaned_weights = ef.clean_weights()
print(cleaned_weights)
ef.portfolio_performance(verbose=True)
```
